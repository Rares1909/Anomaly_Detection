```python
import numpy as np
import matplotlib.pyplot as plt
from pyod.utils.data import generate_data_clusters
from pyod.models.knn import KNN
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from pyod.models.lof import LOF
from sklearn.metrics import confusion_matrix
from scipy.io import loadmat
from pyod.models.combination import average, maximization
from pyod.utils.utility import standardizer
```

```python
def leverage_scores(X):
    U, S, Vh = np.linalg.svd(X, full_matrices=False)
    h=U@U.T
    leverage_scores = np.diag(h)
    return leverage_scores

    
def linear_model(a,b,noise_type,mu,sigma,n_points=100):
    if noise_type == "regular":
        x1 = np.random.uniform(-5, 5, n_points)
        epsilon = np.random.normal(mu, sigma, n_points)
    elif noise_type == "high_x_var":
        x1 = np.random.uniform(-20, 20, n_points)
        epsilon = np.random.normal(mu, sigma, n_points)
    elif noise_type == "high_y_var":
        x1 = np.random.uniform(-5, 5, n_points)
        epsilon = np.random.normal(mu, sigma * 5, n_points)
    elif noise_type == "high_both_var":
        x1 = np.random.uniform(-20, 20, n_points)
        epsilon = np.random.normal(mu, sigma * 5, n_points)
    
    y = a * x1 + b + epsilon
    return x1,y

def linear_model_2D(a,b,c,noise_type,mu,sigma,n_points=100):
    if noise_type == "regular":
        x1,x2 = np.random.uniform(-5, 5, n_points), np.random.uniform(-5, 5, n_points)
        epsilon = np.random.normal(mu, sigma, n_points)
    elif noise_type == "high_x_var":
        x1,x2 = np.random.uniform(-20, 20, n_points), np.random.uniform(-20, 20, n_points)
        epsilon = np.random.normal(mu, sigma, n_points)
    elif noise_type == "high_y_var":
        x1,x2 = np.random.uniform(-5, 5, n_points), np.random.uniform(-5, 5, n_points)
        epsilon = np.random.normal(mu, sigma * 5, n_points)
    elif noise_type == "high_both_var":
        x1,x2 = np.random.uniform(-20, 20, n_points), np.random.uniform(-20, 20, n_points)
        epsilon = np.random.normal(mu, sigma * 5, n_points)
    
    y = a * x1 + b*x2 + c +  epsilon
    return x1,x2,y



def plot(x,y,leverage_scores,d):
    plt.figure(figsize=(12, 6))

    # Plot y vs x
    plt.subplot(1, 2, 1)
    plt.scatter(x, y, label='Data Points', color='blue')
    plt.title(f'{d}_{noise_type}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)

    # Highlight high leverage points
    high_leverage_threshold = 2 * np.mean(leverage_scores)
    high_leverage_indices = np.where(leverage_scores > high_leverage_threshold)[0]
    plt.scatter(x[high_leverage_indices], y[high_leverage_indices], 
                color='red', label='High Leverage Points', edgecolor='black')
    plt.legend()

    plt.tight_layout()
    plt.show()


for noise_type in ["regular", "high_x_var", "high_y_var", "high_both_var"]:
    x,y=linear_model(2,5,noise_type,0,1)

    X = np.vstack([x, np.ones_like(x)]).T
    scores=leverage_scores(X)
    plot(x,y,scores,d="1D")

fig = plt.figure(figsize=(15, 12))

for i, noise_type in enumerate(["regular", "high_x_var", "high_y_var", "high_both_var"]):
    ax = fig.add_subplot(2, 2, i+1, projection='3d')
    x1,x2,y=linear_model_2D(2,5,7,noise_type,0,1)

    X = np.vstack([x1,x2, np.ones_like(x1)]).T
    scores=leverage_scores(X)

    mean_score = np.mean(scores)
    colors = np.where(scores > 2 * mean_score, 'r', 'b')
    
    ax.scatter(x1, x2, y, c=colors, alpha=0.6)
    
    # Customize each subplot
    ax.set_title(f'Noise Type: {noise_type}')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('Y')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()

    
```

```python
def compute_balanced_accuracy(pred, labels):
    tp, fn, fp, tn=confusion_matrix(labels, pred).ravel()
    sensitivity=tp/(tp+fn)
    specifity=tn/(fp+tn)
    balanced_accuracy=(specifity + sensitivity)/2
    return balanced_accuracy
```

```python
x_train, x_test, y_train, y_test=generate_data_clusters(n_train=400,n_test=200,n_clusters=2,n_features=2,contamination=0.1)

def ex2(n):
    knn=KNN(contamination=0.1,n_neighbors=n)
    knn.fit(x_train)
    pred_train=knn.predict(x_train)
    pred_test=knn.predict(x_test)
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))


    axes[0, 0].scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap='bwr', alpha=0.6)
    axes[0, 0].set_title('Ground Truth for Train')
    axes[0, 0].set_xlabel('Feature 1')
    axes[0, 0].set_ylabel('Feature 2')
    

    axes[0, 1].scatter(x_train[:, 0], x_train[:, 1], c=pred_train, cmap='bwr', alpha=0.6)
    axes[0, 1].set_title('Predictions for Train')
    axes[0, 1].set_xlabel('Feature 1')
    axes[0, 1].set_ylabel('Feature 2')
    

    axes[1, 0].scatter(x_test[:, 0], x_test[:, 1], c=y_test, cmap='bwr', alpha=0.6)
    axes[1, 0].set_title('Ground Truth for Test')
    axes[1, 0].set_xlabel('Feature 1')
    axes[1, 0].set_ylabel('Feature 2')
    

    axes[1, 1].scatter(x_test[:, 0], x_test[:, 1], c=pred_test, cmap='bwr', alpha=0.6)
    axes[1, 1].set_title('Predictions for Test')
    axes[1, 1].set_xlabel('Feature 1')
    axes[1, 1].set_ylabel('Feature 2')
    
    fig.suptitle(f"neighbors: {n}", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95]) 
    plt.show()

    return pred_train, pred_test

for n in [5,10,15]:
    pred_train, pred_test=ex2(n)
    ac_tr=compute_balanced_accuracy(pred_train,y_train)
    ac_te=compute_balanced_accuracy(pred_test,y_test)
    print(f"Balanced accuracy for train {ac_tr}, balanced accuracy for test{ac_te}, neighbors {n}")
    
        
```

```python
sample1=make_blobs(n_samples=200,n_features=2,cluster_std=2,center_box=(-10,-10))
sample2=make_blobs(n_samples=100,n_features=2,cluster_std=6,center_box=(10,10))
samples=np.vstack((sample1[0],sample2[0]))
def ex3(n_k,n_l):
    knn=KNN(contamination=0.07,n_neighbors=n_k)
    lof=LOF(contamination=0.07,n_neighbors=n_l)
    knn.fit(samples)
    lof.fit(samples)
    pred_knn=knn.predict(samples)
    pred_lof=lof.predict(samples)

    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    # Plot KNN predictions
    axes[0].scatter(samples[:, 0], samples[:, 1], c=pred_knn, cmap='bwr', alpha=0.6)
    axes[0].set_title('KNN Predictions')
    axes[0].set_xlabel('Feature 1')
    axes[0].set_ylabel('Feature 2')
    
    # Plot LOF predictions
    axes[1].scatter(samples[:, 0], samples[:, 1], c=pred_lof, cmap='bwr', alpha=0.6)
    axes[1].set_title('LOF Predictions')
    axes[1].set_xlabel('Feature 1')
    axes[1].set_ylabel('Feature 2')
    
    # Adjust layout for better viewing
    plt.tight_layout()
    plt.suptitle(f"Neighbors for knn {n_k}, neighbors for lof {n_l}", fontsize=16, y=1.02)
    plt.show()

for n_k,n_l in [[5,10],[10,20],[15,30]]:
    ex3(n_k,n_l)
    
    
```

```python
data=loadmat("cardio.mat")

X_train, X_test, y_train, y_test = train_test_split(data["X"], data["y"], test_size=0.2, random_state=42)
X_train_norm, X_test_norm = standardizer(X_train, X_test)
neighbors=[x for x in range(30,121,10)]
all_train_scores=[]
all_test_scores=[]
for n in neighbors:
    lof=LOF(n_neighbors=n)
    lof.fit(X_train_norm)
    pred_train=lof.predict(X_train_norm)
    pred_test=lof.predict(X_test_norm)
    ba_train=compute_balanced_accuracy(pred_train,y_train)
    ba_test=compute_balanced_accuracy(pred_test, y_test)
    print(f"Balanced accuracy score for neighbors {n} on train {ba_train}, on test {ba_test}")
    scores_train=lof.decision_function(X_train_norm)
    scores_test=lof.decision_function(X_test_norm)
    all_train_scores.append(scores_train)
    all_test_scores.append(scores_test)

all_train_scores=np.array(all_train_scores)
all_test_scores=np.array(all_test_scores)
train_scores_norm, test_scores_norm = standardizer(all_train_scores.T, all_test_scores.T)

avg_train=average(train_scores_norm)
avg_test=average(test_scores_norm)

max_train=maximization(train_scores_norm)
max_test=maximization(test_scores_norm)

outliers_rate=np.sum(y_train==1)/len(y_train)
threshold_avg = np.quantile(avg_train, 1 - outliers_rate)
threshold_max = np.quantile(max_train, 1 - outliers_rate)
pred_avg = (avg_test >= threshold_avg).astype(int)
pred_max = (max_test >= threshold_max).astype(int)
print(f"Balanced accuracy score using the average strategy {compute_balanced_accuracy(pred_avg,y_test)}, using the maximum strategy {compute_balanced_accuracy(pred_max,y_test)}")
```
